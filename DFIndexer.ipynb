```python
import pyterrier as pt
if not pt.started():
  pt.init()
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import re
pd.set_option('display.max_colwidth', 150)
```

```python
df=pd.read_csv('linkdin_Job_data.csv')
df
```

```python
df["job"]=df['job'].astype(str)
```

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
nltk.download('punkt')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
def preprocess(text):
    tokens = word_tokenize(text)
    filtered_tokens = [word.lower() for word in tokens if word.lower() not in stop_words]
    return ' '.join(filtered_tokens)
```

```python
def clean(text):
   text = re.sub(r"http\S+", " ", text)
   text = re.sub(r"RT ", " ", text)
   text = re.sub(r"@[\w]*", " ", text)
   text = re.sub(r"[\.\,\#_\|\:\?\?\/\=]", " ", text)
   text = re.sub(r'\t', ' ', text)
   text = re.sub(r'\n', ' ', text)
   text = re.sub(r"\s+", " ", text)
   text = text.strip()
   return text
df['processed_text'] = df['job'].apply(clean)
df
```

```python
df["processed_text"]=df['job'].apply(preprocess)
df
```

```python
from nltk.stem import *
from nltk.stem.porter import *
```

```python
stemmer = PorterStemmer()
```

```python
def Steem_text(text):

    tokens = word_tokenize(text)
    stemmed_tokens = [stemmer.stem(word) for word in tokens]
    return ' '.join(stemmed_tokens)
```

```python
df['processed_text']=df['processed_text'].apply(Steem_text)
df
```

```python
indexer = pt.DFIndexer("./myFirstIndex", overwrite=True)
index_ref = indexer.index(df["processed_text"], docno=df["job_ID"].astype(str))
index = pt.IndexFactory.of(index_ref)
print(index.getCollectionStatistics().toString())
```

```python
term="AI"
term=stemmer.stem(term)
print("The term after most preprocess",term)
```

```python
index.getLexicon()[term].getDocumentFrequency()     #how many document occurs in it
```

```python
job_Ids = []
# the lexicon entry is also our pointer to access the inverted index posting list
for posting in index.getInvertedIndex().getPostings( index.getLexicon().getLexiconEntry( term ) ):
    docno = index.getMetaIndex().getItem("docno", posting.getId())
    job_Ids.append(docno)

print("The job ID's are :", job_Ids)
```
